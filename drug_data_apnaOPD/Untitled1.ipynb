```python
import requests
```

```python
for generic,brand,manufacturer,types in zip(df['Generic Name'],df['Brand Name'],df['Manufacturer'],df['Type']):
    try:
        medicine = {}
        medicine['generic_name']=generic.replace(u'\xa0',u'')
        medicine['brand_name']= brand.replace(u'\xa0',u'')[1:]
        medicine['manufacturer'] = manufacturer.replace(u'\xa0',u'')
        medicine['type'] = types.replace(u'\xa0',u'')
        r = requests.post("http://13.229.113.41:3000/api/medicines", data=medicine)
        print(r)
    except:
        print(generic)
        print(brand)
        print(manufacturer)
        print(types)
    
```
